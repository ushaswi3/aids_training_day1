```python
'''insights:
from barplots, we can see that the mean and std of yellow is greater than green


Right-skewed distribution
Most tips are small (between 0 and ~4).
Very few customers give high tips (>10).
This is typical real-world tipping behavior.
Peak at low tip values
The highest frequency is near 0–1.
Yellow group dominates
Yellow bars are much higher, meaning:
More observations
Or higher frequency of tips in this category
This group contributes most of the overall pattern.
Green group is smaller & concentrated
Green distribution is:
Less frequent
Mostly concentrated at lower tip values
Suggests fewer high tips in this group.
Long tail
Tips go up to around 22–23, but these are rare outliers.
A few generous customers increase the maximum value.
'''
```

```python
# Independent T-Test
yellow_tip = df['tip'][df['color']=='yellow']
green_tip = df['tip'][df['color']=='green']
```

```python
from scipy import stats

t_stat, p_val = stats.ttest_ind(yellow_tip, green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python
alpha=0.05
if p_val < alpha:
    print("Reject null hypothesis: The tip for yellow category is different that of green")
else:
    print("Fail to reject null hypothesis")

#There is effect of color on tip column
```
