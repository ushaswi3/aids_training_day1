```python
'''insights:
from barplots, we can see that the mean and std of yellow is greater than green


Right-skewed distribution
Most tips are small (between 0 and ~4).
Very few customers give high tips (>10).
This is typical real-world tipping behavior.
Peak at low tip values
The highest frequency is near 0–1.
Yellow group dominates
Yellow bars are much higher, meaning:
More observations
Or higher frequency of tips in this category
This group contributes most of the overall pattern.
Green group is smaller & concentrated
Green distribution is:
Less frequent
Mostly concentrated at lower tip values
Suggests fewer high tips in this group.
Long tail
Tips go up to around 22–23, but these are rare outliers.
A few generous customers increase the maximum value.
'''
```

```python
# Independent T-Test
yellow_tip = df['tip'][df['color']=='yellow']
green_tip = df['tip'][df['color']=='green']
```

```python
from scipy import stats

t_stat, p_val = stats.ttest_ind(yellow_tip, green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python
alpha=0.05
if p_val < alpha:
    print("Reject null hypothesis: The tip for yellow category is different that of green")
else:
    print("Fail to reject null hypothesis")

#There is effect of color on tip column
```

```python
"Hypothesis:"
'''
-H0 : distribution of tip is the same across all pickup_brough categories
-H1 : At least one category has a different distribution of tip

'''
```

```python
Manhattan_tip=df['tip'][df['pickup_borough']=="Manhattan"]
Queens_tip=df['tip'][df['pickup_borough']=="Queens"]
Bronx_tip=df['tip'][df['pickup_borough']=="Bronx"]
Brooklyn_tip=df['tip'][df['pickup_borough']=="Brooklyn"]
```

```python
# ANOVA Test

f_stat, p_val = stats.f_oneway(Manhattan_tip,Queens_tip,Bronx_tip,Brooklyn_tip)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

alpha=0.05
if p_val < alpha:
    print("Reject null hypothesis: Spending Score differs by Region")
else:
    print("Fail to reject null hypothesis: No significant difference by Region")
```

```python
# Tukey HSD Test
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform Tukey HSD
tukey_result = pairwise_tukeyhsd(endog=df.tip, groups=df.pickup_borough, alpha=0.05)
print(tukey_result)
```

```python
"Hypothesis:"
'''
-H0 : distribution of tip is the same across all dropoff_brough categories
-H1 : At least one category has a different distribution of tip

'''
```

```python
dropoff_m=df[df['dropoff_borough']=='Manhattan']['tip']
dropoff_q=df[df['dropoff_borough']=='Queens']['tip']
dropoff_bx=df[df['dropoff_borough']=='Bronx']['tip']
dropoff_bn=df[df['dropoff_borough']=='Brooklyn']['tip']
dropoff_s=df[df['dropoff_borough']=='Staten Island']['tip']

f_stat_drop,p_val_drop=stats.f_oneway(dropoff_m,dropoff_q,dropoff_bx,dropoff_bn,dropoff_s)
print("ANOVA F-statistic:", f_stat_drop)
print("P-value:", p_val_drop)

if(p_val_drop<0.05):
    print('reject null hypothesis : tip of taxi customer is differ by dropoff_borough of taxi')
else:
    print('fail to reject null hypothesis : No significant tip difference by dropoff_borough')
```

```python
# Anova Test
categories=[group['tip'].values for name, group in df.groupby('pickup_zone')]

f_stat_drop,p_val_drop=stats.f_oneway(*categories)
print("ANOVA F-statistic:", f_stat_drop)
print("P-value:", p_val_drop)

if(p_val_drop<0.05):
    print('reject null hypothesis ')
else:
    print('fail to reject null hypothesis ')
```

```python
# Anova Test
dropoff_categories=[group['tip'].values for name, group in df.groupby('dropoff_zone')]

f_stat_drop,p_val_drop=stats.f_oneway(*dropoff_categories)
print("ANOVA F-statistic:", f_stat_drop)
print("P-value:", p_val_drop)

if(p_val_drop<0.05):
    print('reject null hypothesis ')
else:
    print('fail to reject null hypothesis ')
```
