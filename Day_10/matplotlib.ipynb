```python
'''•  Inspect the dataset and list the number of rows, columns, data types, and any missing values.
•  Remove duplicate rows (if any) and re-check the dataset — how many rows were removed?
•  Parse the Date column to datetime, sort the data by date, and report how many rows (if any) have invalid/missing dates.
•  Handle missing values in numeric columns (explain and apply a method you choose), then state which method you used and why.'''

from google.colab import drive
import pandas as pd

#Mount your Drive
drive.mount('/content/drive')

#once mounted,navigate to the folder
#example path(adjust bsed on your structure)
path='/content/drive/My Drive/Colab Notebooks/data.csv'

df=pd.read_csv(path)
```

```python
'''  Inspect the dataset and list the number of rows,
columns, data types, and any missing values. '''
import pandas as pd



# Inspect dataset
print("Number of rows and columns:", df.shape)
print("\nData types:\n", df.dtypes)
print("\nMissing values per column:\n", df.isnull().sum())

print(df.isnull())
print(df.isnull().sum())

'''Remove duplicate rows (if any) and re-check the dataset —
how many rows were removed?'''
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
df = df.drop_duplicates()
print("after removing duplicates:", df.shape[0])
print("Rows removed:", duplicates)

'''  Parse the Date column to datetime, sort the data by date,
 and report how many rows (if any) have invalid/missing dates. '''
# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Sort data by date
df = df.sort_values(by="Date")

# Check invalid/missing dates
invalid_dates = df["Date"].isna().sum()
print("Number of invalid/missing Date entries:", invalid_dates)

'''Handle missing values in numeric columns (explain and apply a method you choose),
 then state which method you used and why. '''
 # Select numeric columns
num_cols = ["Duration", "Pulse", "Maxpulse", "Calories"]

# Check which have missing values
print(df[num_cols].isnull().sum())

# Example method: Fill missing numeric values with column mean
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Recheck
print("\nAfter imputation:\n", df[num_cols].isnull().sum())


```

```python
'''Create a line plot of Calories over Date
(with markers, dashed line, grid, rotated x-ticks)
and save the image as line_calories_over_time.png.'''
import matplotlib.pyplot as plt
cal=df['Calories']
date=df['Date']
#by default it takes line plot
plt.plot(cal,date,color='blue',marker='o',linestyle='--')
#Add lables and title
plt.title("Calories over Date")
plt.xlabel("Calories")
plt.ylabel("Date")
plt.grid(True)
plt.show()
```

```python
'''Group data by Duration (or sensible duration bins),
compute the average Pulse per group, and draw a bar chart
comparing average pulse across duration groups;
save as bar_avg_pulse_by_duration.png. '''


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
avg_pulse_by_duration = df.groupby("Duration")["Pulse"].mean()
bins = [0, 30, 60, 90, 120, 150]
labels = ['0-30', '31-60', '61-90', '91-120', '121-150']
df['Duration_bin'] = pd.cut(df['Duration'], bins=bins, labels=labels, include_lowest=True)
avg_pulse_by_bin = df.groupby('Duration_bin')['Pulse'].mean()

plt.figure(figsize=(8,5))
avg_pulse_by_bin.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Pulse by Duration Group")
plt.xlabel("Duration (minutes)")
plt.ylabel("Average Pulse")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("bar_avg_pulse_by_duration.png")
plt.show()
```

```python
'''   Plot a scatter plot of Duration vs Calories,
compute the Pearson correlation coefficient,
and add a linear trend line to the scatter;
save as scatter_duration_vs_calories.png.'''

import matplotlib.pyplot as plt
import numpy as np

duration = df['Duration']
cal = df['Calories']

plt.scatter(duration, cal, color='blue')
m, b = np.polyfit(duration, cal, 1)
plt.plot(duration, m * duration + b, color='red')
corr = np.corrcoef(duration, cal)[0, 1]
plt.title(f"Duration vs Calories (r = {corr:.2f})")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.tight_layout()
plt.savefig("scatter_duration_vs_calories.png")
plt.show()

```

```python
'''  Create a pie chart showing the percentage breakdown of Maxpulse into
ranges (e.g., <=120, 121–140, 141–160, >160) and
save as pie_maxpulse_ranges.png.'''

import matplotlib.pyplot as plt
import pandas as pd

max_p = df['Maxpulse']

labels = ['<=120', '121-140', '141-160', '>160']
sizes = [
    (max_p <= 120).sum(),
    ((max_p > 120) & (max_p <= 140)).sum(),
    ((max_p > 140) & (max_p <= 160)).sum(),
    (max_p > 160).sum()
]
colors = ['green', 'yellow', 'blue', 'grey']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Percentage Breakdown of Maxpulse Ranges")
plt.tight_layout()
plt.savefig("pie_maxpulse_ranges.png")
plt.show()

```

```python
'''Plot a histogram of Calories with an appropriate number of bins,
comment on shape/skewness, and save as hist_calories.png.'''

cal=df['Calories']
plt.hist(cal,bins=8,color='lightblue',edgecolor='black')
plt.title("Histogram of Calories")
plt.xlabel("Calories")
plt.ylabel("Count")
plt.show()
```

```python
'''  Build a multi-panel subplot figure (1 row × 2 columns):
left = line plot of Pulse over Date, right = scatter of Duration vs Maxpulse;
save as subplots_pulse_duration_maxpulse.png.'''

import matplotlib.pyplot as plt
import pandas as pd

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.sort_values('Date')

fig, axes = plt.subplots(1, 2, figsize=(14,5))

axes[0].plot(df['Date'], df['Pulse'], color='blue')
axes[0].set_title("Pulse over Date")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Pulse")

axes[1].scatter(df['Duration'], df['Maxpulse'], color='red')
axes[1].set_title("Duration vs Maxpulse")
axes[1].set_xlabel("Duration")
axes[1].set_ylabel("Maxpulse")

plt.tight_layout()
plt.savefig("subplots_pulse_duration_maxpulse.png")
plt.show()
```
